```python
import json
from datasets import Dataset

INPUT_PATH = "tiki_training_dataset.json"
OUTPUT_PATH = "tiki_dataset_processed.jsonl"

def main():
    with open(INPUT_PATH, "r", encoding="utf-8") as f:
        raw_data = json.load(f)

    processed = []
    for item in raw_data:
        system_prompt = ""
        user_msg = ""
        assistant_msg = ""

        for msg in item["messages"]:
            if msg["role"] == "system":
                system_prompt = msg["content"].strip()
            elif msg["role"] == "user":
                user_msg = msg["content"].strip()
            elif msg["role"] == "assistant":
                assistant_msg = msg["content"].strip()

        if user_msg and assistant_msg:
            full_text = (
                f"<|system|>: {system_prompt}\n"
                f"<|user|>: {user_msg}\n"
                f"<|assistant|>: {assistant_msg}"
            )
            processed.append({
                "system": system_prompt,
                "instruction": user_msg,
                "output": assistant_msg,
                "text": full_text
            })

    with open(OUTPUT_PATH, "w", encoding="utf-8") as f:
        for row in processed:
            json.dump(row, f, ensure_ascii=False)
            f.write("\n")

    print(f"Đã xử lý {len(processed)} mẫu và lưu vào {OUTPUT_PATH}")

if __name__ == "__main__":
    main()
```
